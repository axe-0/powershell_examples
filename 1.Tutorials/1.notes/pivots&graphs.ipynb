# Data Visualisation

```python
#read csv file
df_population_raw = pd.read_csv('/Users/xolo/Desktop/Python Data analysis/2.Python for Excel Users/population_total.csv')
df_population_raw
```

## Maving a Pivot Table


```python
#dropping null values 
df_population_raw = df_population_raw.dropna()
df_population_raw
```

```python
# List of countries you want to include
countries_to_include = ['China', 'Zambia', 'India', 'Andorra', 'Yemen', 'Algeria']

# Filter the DataFrame to include only the specified countries
df_filtered = df_population_raw[df_population_raw['country'].isin(countries_to_include)]

# Pivot the filtered DataFrame
df_pop_pivot = df_filtered.pivot(index='country', columns='year', values='population')
```

```python
#select columns 
df_pop_pivot
```

```python
df_pop_transposed = df_pop_pivot.transpose()
df_pop_transposed 
```

### 1. Line Plot

```python
#Line plot 
df_pop_transposed .plot(kind = 'line', xlabel = 'Year', ylabel = 'Population', title = 'Population (1955-2020)', figsize= (8,4))

#Save plot
plt.savefig('line_plot.png')

#show plot 
plt.show()
```

### 2.1 Bar plot

```python
#selecting a year (2020)
df_pop_2020 = df_pop_transposed[df_pop_transposed.index.isin([2020])]
df_pop_2020 = df_pop_2020.T
df_pop_2020
```

```python
#make a barplot 
df_pop_2020.plot(kind = 'bar', xlabel = 'Country', ylabel = 'Population', title = 'Population (2020)',color = 'orange')

#Save plot 
plt.savefig('bar_plot.png')

#show plot
plt.show()
```

### 2.2 Bar Plot grouped by n variables 

```python
#selecting a few years 
df_pop_5yr = df_pop_transposed[df_pop_transposed.index.isin([2015,2016,2017,2018,2019,2020])]
df_pop_5yr 
```

```python
#making grouped barplot 
df_pop_5yr.plot(kind = 'bar',xlabel = 'Years', ylabel = 'Population', title = 'Population (2015- 2020)')

#Save plot 
plt.savefig('group_barplot.png')

#show plot
plt.show()
```

### 3.Pie Chart 

```python
#changing column name 
df_popu_2020 =df_pop_transposed[df_pop_transposed.index.isin([2020])]
df_popu_2020 = df_popu_2020.T
df_popu_n2020 =df_popu_2020.rename(columns= {2020:'2020'})
df_popu_n2020
```

```python
df_popu_n2020.plot(kind = 'pie',y ='2020', title = 'Population in 2020 (%)')

#Save plot 
plt.savefig('pie_chart.png')

#show plot
plt.show()
```
